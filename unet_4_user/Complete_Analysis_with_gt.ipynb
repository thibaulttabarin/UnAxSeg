# Analysis pipeline for LM image using unet : image to axon/myeling statistic
This pipeline generate the prediction of the mask for a RGB image (using a convertion) and apply the myelin segmentation then create the statistics with corresponding plot.<br>
This analysis requires the ground truth data
1. Import raw image and the model
2. Apply the model to image 
3. Myelin instance segmentation 
4. Create a dataframe
5. Statistic and plots
6. If you only have ground truth

```python
import os
from PIL import Image
from PIL import ImageOps
import numpy as np
import skimage.io as io
import math
import matplotlib.pyplot as plt
from keras.models import load_model
```

```python
import unet_4_user.utility.Performance_measurement as Perf_m
import unet_4_user.utility.Apply_model as Apply_model
from unet_4_user.utility.Utility import Image_2_np, Import_image
import unet_4_user.utility.Myelin_segmentation_utility as ms_uti
from unet_4_user.utility.utility_plot import *
```

1- Import raw data and model

```python
# import model and image
# To choose the image or model either change the path below or make the path empty('') 
# and it will open a dialog box
model_path = './Models/model_in_use.h5'
image_path = './Data/dataset_demo/Train_Validation/sample1/image.png'
model = Apply_model.Import_model(model_path = model_path)
img, imgage_path = Import_image(image_path)
```

2- Apply model

```python
# Apply the model
pred_image = Apply_model.Unet_by_patches(img, model, patch_size = 256, overlap=64)

mask = np.argmax(pred_image, axis=-1)
mask=127*mask.astype(np.uint8)
```

```python
# Sanity check
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(15,10))
ax[0].imshow(np.array(img))
ax[1].imshow(mask)
```

```python
# save the image either as numpy array or as image or skip this part
folder_path,_ = os.path.split(image_path)
np.save(folder_path + '/unet_np_array', mask)
io.imsave(folder_path+ '/unet_image.png', mask)
```

## 3- Myelin instance segmentation


### Myelin Segmentation of the predicted mask

```python
# Segment the fibers
fiber_pred_seg = ms_uti.quick_myelin_seg(mask)
_, _, label_fiber_pred, label_axon_pred  = ms_uti.clean_up (fiber_pred_seg, mask)
del fiber_pred_seg, mask
```

```python
# Sanity check
# Display the segmentation of the fiber
label_myelin_pred = label_fiber_pred-label_axon_pred
mycolor=ms_uti.random_cmap(label_fiber_pred.max(),'cool')
fig, ax =  plt.subplots(figsize = (18,5),nrows=1, ncols=3, sharex=True, sharey=True)
fig.suptitle('Instance Segmentation', fontsize = 16)
ax=ax.flatten()
ax[0].imshow(label_fiber_pred, cmap =mycolor)
ax[0].set_title('fiber pred')
ax[1].imshow(label_myelin_pred, cmap =mycolor)
ax[1].set_title('myelin pred')
ax[2].imshow(label_axon_pred, cmap =mycolor)
ax[2].set_title('axon pred')
```

### Myelin Segmentation of the ground mask¶

```python
# load the ground_truth mask if relevant
# and segment the fiber
#print(folder_path)
gt_mask, _ = Image_2_np(title='select ground truth mask')

fiber_gt_seg = ms_uti.quick_myelin_seg(gt_mask)
_, _, label_fiber_gt, label_axon_gt  = ms_uti.clean_up (fiber_gt_seg, gt_mask)
del  fiber_gt_seg
```

```python
# Sanity check
# Display the segmentation of the fiber
label_myelin_gt = label_fiber_gt-label_axon_gt
mycolor=ms_uti.random_cmap(label_fiber_gt.max(),'cool')
fig, ax =  plt.subplots(figsize = (18,5),nrows=1, ncols=3, sharex=True, sharey=True)
fig.suptitle('Instance Segmentation', fontsize = 16)
ax=ax.flatten()
ax[0].imshow(label_fiber_gt, cmap =mycolor)
ax[0].set_title('fiber gt')
ax[1].imshow(label_myelin_gt, cmap =mycolor)
ax[1].set_title('myelin gt')
ax[2].imshow(label_axon_gt, cmap =mycolor)
ax[2].set_title('axon gt')
```

### Plot 2 : Bland-altman
- Plot the bland-altman plot.
- Draw threshold lines at -/+ 1.96 of standart deviation (1.96sd) representing the 95% of confidence intervale.
- Indicate fiber in(blue) and out(red) of the confidence intervale.
- Indicate the False Positive fiber in green (means and difference both equal to the predicted value).

```python
# Plot 2 Bland-altman plot for fiber threshold a -+1.96sd
title_2 = 'Bland-altman plot :\n predicted vs true {0} {1}'.format(measured, unit)

pred_xx_TP =df.loc[df['true_positive']==1, pred]
pred_xx_FP =df.loc[df['true_positive']==0, pred]

gt_xx_TP =df.loc[df['true_positive']==1, gt]
gt_xx_FP =df.loc[df['true_positive']==0, gt]

plt.figure()
_, (lower, upper)= bland_altman_plot_2(gt_xx_TP, pred_xx_TP, res_limit= None, sd_limit= 1.96)
plt.scatter(pred_xx_FP, pred_xx_FP, color='green')
plt.title(title_2, fontsize=16)
# if x and y axis need scaling
plt.xlim(-0.1, 1.5)
plt.ylim(-0.6, 1.5)
plt.tight_layout()
```

### Plot 3 : Predicted vs ground truth + linear fit
- Plot predicted value vs ground truth.
- fit a linear model display the fit green line and display fitting parameter
- Display the y=x model (as a comparison with) 

```python
# plot 3 : Predicted diameter vs true diameter + linear fit
x_label_3 = 'True {0} {1}'.format(measured, unit )
y_label_3 = 'Predicted {0} {1}'.format(measured, unit )
title_3 = '{0} True positive:\n true vs predicted'. format(measured)

x= pred_xx_TP
pred_y= gt_xx_TP

slope, intercept, r_value, p_value, std_err = stats.linregress(x,pred_y)
fit_y = intercept + slope*x

plt.figure()
plt.scatter(x, pred_y)
plt.plot(x, x, color='red')
plt.plot(x, fit_y, color='green', linestyle = '--')
plt.title(title_3,fontsize=16)
plt.xlabel(x_label_3,fontsize=16); plt.ylabel(y_label_3,fontsize=16)
legend_fit = 'y=%.2f +%.2f  \n R-squared : %.2f ' % (intercept, slope,r_value**2)
plt.legend(['y=x', legend_fit, 'data'])
plt.tight_layout()
```

### Plot 5 : for the feature of interest percentage per bin  of out 1.96sd and FP
- Directly relate to the plot 4 
- Display categories : out(red) confidence interval (1.96sd) and False Positive(green)

```python
title_5 = '% out and FP per bin \n(out of total predicted fibers)'
x_label_5 = 'Predicted {0} {1}'.format(measured, unit )
y_label_5 = 'Percentage per bin'

df ['diff_'] =df[gt]-df[pred]
xx= df[pred]
xx_out_sd = df.loc[((df['diff_']<lower) | (df['diff_']>upper)) & (df['true_positive']==1), pred]
xx_FP = df.loc[df['true_positive']==0, pred]

n_xx, bins =np.histogram(xx, bins=50)
n_xx_out_sd, _ = np.histogram(xx_out_sd, bins=bins)
n_xx_FP, _ = np.histogram(xx_FP, bins=bins)

bin_width = (bins[1] - bins[0])
bin_centers = bins[1:] - bin_width/2
plt.figure()
b1= plt.bar(bin_centers, 100*n_xx_out_sd/n_xx, bin_width, color='red', edgecolor='black')
b2= plt.bar(bin_centers, 100*n_xx_FP/n_xx, bin_width, color='green',bottom=100*n_xx_out_sd/n_xx, edgecolor='black')
legend1 = '% out per bin'
legend2 = '% FP per bin'
plt.xlabel(x_label_1, fontsize=16); plt.ylabel(y_label_1, fontsize=16)
plt.title(title_1, fontsize=16)
plt.legend([b1,b2], [legend1,legend2], fontsize=16, loc='upper right')
plt.xlabel(x_label_5, fontsize=16); plt.ylabel(y_label_5,fontsize=16)
plt.title(title_5, fontsize=16)
plt.tight_layout()
```
