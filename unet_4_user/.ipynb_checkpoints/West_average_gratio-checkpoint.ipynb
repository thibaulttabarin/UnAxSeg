# West et al. revised model for estimating gratio
MRI and histology data weighted gratio
Using the West et al. A resived model for estimated g-ratio from MRI
The model propose by West to calculate g ratio using volume fraction (of axon, myelin and fiber) use the area weighted mean square of the gratio to approcximate the MRI gratio.
In our case it seems that the distribution of gratio display to peak population of gratio, onee around .35 and 0.62. The MRI gratio seems to capture solely the estimate of the high gratio population corresponding to the larger axons, the awm gratio capture a very similar value. Although we have a much larger population of small axon.
Should we design simulation to assess the impact of a multimodal distribution of the different measure of the gratio.

## 2- Create the mask

```python
# Create the mask
mask =  np.argmax(prediction, axis=2)

Myelin_mask = mask == 1
Axon_mask = mask == 2
mask_ = np.zeros_like(mask, dtype=np.uint8)
mask_[Myelin_mask]=127
mask_[Axon_mask]=255

fig, ax =plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize=(18,15))
ax[0].imshow(mask_)
ax[0].axis('off')
ax[1].imshow(Myelin_mask)
ax[1].axis('off')
ax[2].imshow(Axon_mask)
ax[2].axis('off')

plt.show()
```

## 3- Segment Myelin
output of the unet is 3 class bkg myelin and axon. Unfortunetally, the myelin are not segemented by fiber.
This part will perform instance segmentation of the myelin in order word segmentation of myelin per fiber

```python
# Segment Myelin
import utility.Myelin_segmentation_utility as ms_uti

# segment
fiber_label = ms_uti.quick_myelin_seg(mask_)
# Clean
_, _, label_fiber_pred, label_axon_pred  = ms_uti.clean_up (fiber_label, mask_)

# Display the segmentation of the fiber
# more compact (may not be necessary)
_ = ms_uti.Fiber_Myelin_Axon_colorcoded(label_fiber_pred, label_axon_pred, figsize=(18,15))
```

```python
#  Create dataframe for the fibers properties features : ['area_f', 'area_a','diameter_f', 'diameter_a', 'gratio']
df_fiber=Perf_m.Fiber_morpho(label_fiber_pred, label_axon_pred)
```

```python
df_fiber.keys()
```

```python
av_gratio = np.mean(df_fiber['gratio'])
```

```python
weight_gratio = np.sqrt(np.sum(df_fiber['area_f']*np.square(df_fiber['gratio'])/np.sum(df_fiber['area_f'])))
```

```python
mask_size = Myelin_mask.shape
Total = mask_size[0]*mask_size[1]
AVF = np.sum(Axon_mask)/Total
MVF = np.sum(Myelin_mask)/Total
FVF =AVF+MVF
MRI_gratio = np.sqrt(1-MVF/FVF)
```

```python
print('av_gratio : {0:.2f}, weighted_gratio : {1:.2f}, MRI_gratio : {2:.2f}'.format(av_gratio, weight_gratio, MRI_gratio))
```

```python
# Display plot
plt.figure()
ax= df_fiber['gratio'].hist(bins=50, color = 'gray',edgecolor='black')
b1= ax.axvline(av_gratio, color = 'r')  # draw mean line.
b2= ax.axvline(weight_gratio, color = 'g')  # draw mean line.
b3 = ax.axvline(MRI_gratio, color = 'b')  # draw mean line.
ax.set_title('Distribution g ratio ({0} fiber)'.format(len(df_fiber)))
ax.set_xlabel('gratio'); ax.set_ylabel('frequency')
    
legend1 = 'mean : {0:.3f}'.format(av_gratio)
legend2 = 'weighted : {0:.3f}'.format(weight_gratio)
legend3 = 'MRI : {0:.3f}'.format(MRI_gratio)
plt.legend([b1,b2,b3], [legend1, legend2, legend3], fontsize=10)
```

```python
import scipy.stats
data_diam_a = np.array(df_fiber['diam_a'])*0.11
plt.hist(data_diam_a, density=True, bins=50, edgecolor='black')
plt.title('Distribution diameter ({0} fiber)'.format(len(df_fiber)))
plt.xlabel('diameter ${\mu}m$'); plt.ylabel('frequency')
plt.show()
```

## 4- Distribution and pdf fit

```python
import scipy.stats
data_gratio = np.array(df_fiber['gratio'])
hist = np.histogram(data_gratio, bins=50)
hist_dist = scipy.stats.rv_histogram(hist)
```

```python
X = np.linspace(0, 1, 100)
plt.title("PDF from Template")
plt.hist(data_gratio, density=True, bins=50)
plt.plot(X, hist_dist.pdf(X), label='PDF')
plt.plot(X, hist_dist.cdf(X), label='CDF')
plt.title('PDF and CDF gratio')
plt.xlabel('gratio'); plt.ylabel('frequency')
plt.show()
```
